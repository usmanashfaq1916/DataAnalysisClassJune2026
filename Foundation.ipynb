```python
#Talke Two numbers from user and take input from user which operation to perfrom
# Check if results is even or odd and print the result
n1 = int(input("Enter Number 1: "))
n2 = int(input("Enter Number 2: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
    result = n1 + n2
    if result % 2 == 0:
        print(f"The result of {n1} + {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} + {n2} = {result} and it is odd")
elif operation == "-":
    result = n1 - n2    
    if result % 2 == 0:
        print(f"The result of {n1} - {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} - {n2} = {result} and it is odd")

elif operation == "*":
    result = n1 * n2
    if result % 2 == 0:
        print(f"The result of {n1} * {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} * {n2} = {result} and it is odd")
elif operation == "/":
    result = n1 / n2
    if int(result) % 2 == 0:
        print(f"The result of {n1} / {n2} = {result} and it is even")
    else:
        print(f"The result of {n1} / {n2} = {result} and it is odd")
else:
    print("Invalid operation")
    result = None
```
